````python
def full_display(df):
    with pd.option_context('display.max_rows', 1000):
        with pd.option_context('display.max_columns', 1000):
            display(df)
````

````python
def data_quality(df):
    '''
    DESCRIPTION: Displays the quality of data
     
    INPUT: Dataframe
    
    OUTPUT: Returns the total no of null values, percentage null values, data type etc.
    '''
    dtypes = df.dtypes
    unique = df.T.apply(lambda x: x.nunique(), axis=1)
    null = df.isna().sum().sort_values(ascending=False)
    percent = (df.isna().sum()/df.isna().count()*100).sort_values(ascending=False)
    df_quality = pd.concat([null,percent,unique,dtypes],axis=1,keys=['no_of_nulls','%percent_null','unique_items','dtype'])
    display(df_quality)
````

Now for some feature engineering before the filling of nan values. I am taking this route so as to preserve the integrity of the data.

##### Feature Engineering

````python
 mapper = { "K > 24 month": 24, 
          "I 18-21 month": 18, 
          "G 12-15 month": 12, 
          "H 15-18 month": 15, 
          "J 21-24 month": 21, 
          "F 9-12 month": 9, 
          "D 3-6 month": 3, 
          "E 6-9 month": 6   }

train['TENURE_'] = train['TENURE'].replace(mapper)
````

````python
train['TENURE/REGULARITY'] = train['TENURE_'] / train['REGULARITY']
train['TENURE_REGULARITY'] = train['TENURE_'] * train['REGULARITY']
train['REVENUE-MONTANT'] = train['REVENUE'] - train['MONTANT']
train['REVENUE/MONTANT'] = train['REVENUE'] / train['MONTANT']
train['TENURE/FREQUENCE_RECH'] = train['TENURE_'] / train['FREQUENCE_RECH']
````

````python
train['not_expresso'] = train['ORANGE']+train['TIGO']
train['Total_net_calls'] = train['ON_NET']+train['ORANGE']+train['TIGO']
train['%_expresso_calls'] = (train['ON_NET']/train['Total_net_calls'])*100
train['%_not_expresso_calls'] = 1-train['%_expresso_calls']
train['unit_tp'] = train['MONTANT']/train['FREQUENCE_RECH']
````

````python
full_display(train.head())
````

````python
#Total number of missing values
train.isna().sum()
````

##### Filling missing values


* fill numerical features with `-999`, which is quite distinct from all other values so as to keep the integrity of the data.
* fill the categorical features with `NONE` 

````python
categorical_feat = train.select_dtypes(exclude = np.number).columns.to_list()
numerical_feat = train.select_dtypes(exclude = object).columns.to_list()
numerical_feat.remove('CHURN')

for col in numerical_feat:
    train[col].fillna(-999, inplace = True)
for col in categorical_feat:
    train[col].fillna(col+'_NONE', inplace = True)
````

###### more feature engineering

````python
for col in numerical_feat:
    train[col + '_mean_on_reg'] = train[col].groupby(train['REGION']).transform('mean')
    train[col + '_mean_on_ten'] = train[col].groupby(train['TENURE']).transform('mean')
    
    train[col + '_sum_on_reg'] = train[col].groupby(train['REGION']).transform('sum')
    train[col + '_sum_on_ten'] = train[col].groupby(train['TENURE']).transform('sum')
````

checking the distribution of the categorical variables after filling missing values

````python
#countplot
for i in ['TENURE','REGION']:
    print(train[i].value_counts())
    sns.countplot(train[i])
    plt.xticks(rotation=90)
    plt.show()
````

* We can One hot encode ```REGION``` and ```TENURE``` then Label Encode ```TOP_PACK```

we label encode TOP_PACK because it has many unique values, while others have few.

````python
train['TOP_PACK']  = pd.factorize(train['TOP_PACK'])[0]
train = pd.get_dummies(columns= ['REGION', 'TENURE'], data = train)
````

````python
from sklearn.model_selection import train_test_split,StratifiedKFold, GridSearchCV
from sklearn.metrics import log_loss
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from catboost import CatBoostClassifier
````

````python
y=train['CHURN']
train.drop(['user_id','CHURN'],axis=1, inplace=True)

X_train, X_test, y_train, y_test = train_test_split(train, y, test_size=0.3,stratify = y,  random_state=42)
````

````python
lr = LogisticRegression()
lr.fit(X_train, y_train)
print('Logistic Regression Log loss: {}'.format(log_loss(y_test, lr.predict_proba(X_test)[:,1])))
````

````python
cat = CatBoostClassifier(
    n_estimators= 4000,
    learning_rate= 0.01,
    rsm= 0.7)
cat.fit(X_train, y_train)
````

````python
print('catboost Log loss: {}'.format(log_loss(y_test, cat.predict_proba(X_test)[:,1])))
````

````python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
````

````python
print('RandomForestClassifier Log loss: {}'.format(log_loss(y_test, rf.predict_proba(X_test)[:,1])))
````

##### FEATURES IMPORTANCE PLOT

````python
feature_importance_df = pd.DataFrame(rf.feature_importances_, columns=['importance'])
feature_importance_df['feature'] = train.columns

plt.figure(figsize=(12, 6));
sns.barplot(x="importance", y="feature", data=feature_importance_df.sort_values(by = ['importance'], ascending = False).head(50))
plt.title('RandomForestClassifier features importance:');
````

Will be using the catboost classifier, since it gives us the lowest logloss error
